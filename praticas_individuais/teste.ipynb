```python
import pandas as pd
import os

# Diretório onde os arquivos estão localizados
diretorio = '/home/ABTLUS/pedro220058/Documentos/Quinto_Semestre/Computacao_Avancada/dados pca'

# Lista para armazenar os DataFrames de cada arquivo
dataframes = []

# Iterar sobre os arquivos no diretório e subdiretórios
for root, dirs, files in os.walk(diretorio):
    for file in files:
        filename, extension = os.path.splitext(file)
        if extension == '.xls':
            # Caminho completo do arquivo
            caminho_arquivo = os.path.join(root, file)
            # Ler o arquivo Excel em um DataFrame
            df = pd.read_excel(caminho_arquivo)
            # Adicionar o DataFrame à lista
            dataframes.append(df)

# Agora a lista dataframes contém DataFrames correspondentes a todos os arquivos Excel (.xls) no diretório especificado

# Criar 9 variáveis DataFrame df1, df2, df3, ..., df9
df1, df2, df3, df4, df5, df6, df7, df8, df9 = dataframes[:9]
```

```python
# Combine os 9 DataFrames em um único DataFrame
combined_df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9], ignore_index=True)

# Exemplo de visualização das primeiras linhas do DataFrame combinado
combined_df
```

```python
import matplotlib.pyplot as plt

# Contagem de ocorrências de cada idioma na coluna "language"
language_counts = combined_df['Language'].value_counts()

# Criar um gráfico de setores
plt.figure(figsize=(8, 8))
plt.pie(language_counts, labels=language_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Idiomas')
plt.axis('equal')  # Faz com que o gráfico de setores seja circular
plt.tight_layout()  # Ajusta o layout para evitar cortar rótulos
plt.show()
```

```python
import matplotlib.pyplot as plt

# Contagem de ocorrências de cada ano na coluna "Publication Year"
language_counts = combined_df['Publication Year'].value_counts()

# Ordenar os valores pela chave (ano)
language_counts = language_counts.sort_index()

# Criar um gráfico de barras com a cor roxa
plt.figure(figsize=(10, 6))
language_counts.plot(kind='bar', color='purple')  # Mudar a cor para roxo
plt.title('Ano de Publicação')
plt.xlabel('Ano')
plt.ylabel('Número de Artigos')
plt.xticks(rotation=45)  # Rotacionar os rótulos do eixo x para melhor legibilidade
plt.tight_layout()  # Ajustar layout para evitar cortar rótulos
plt.show()
```

```python
import matplotlib.pyplot as plt

# Contagem de ocorrências de cada idioma na coluna "Language"
language_counts = combined_df['Language'].value_counts()

# Criar um gráfico de setores
plt.figure(figsize=(8, 8))
patches, texts, _ = plt.pie(language_counts, labels=language_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribuição de Idiomas')

# Ajuste de layout e tamanho da fonte da legenda
plt.legend(language_counts.index, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
for text in texts:
    text.set_fontsize(10)

plt.axis('equal')  # Faz com que o gráfico de setores seja circular
plt.tight_layout()  # Ajusta o layout para evitar cortar rótulos
plt.show()
```
