## Parsing

```python
# Importação dos abstracts para a execução do parsing

abstracts = []

for i,arq in enumerate(arqs):
    
    if i == 0:
        df = pd.read_excel(f'dados pca/{arq}')
        df = df.loc[df['Language'] == 'English']
        abstracts.extend(df['Abstract'].tolist())
    else:
        aux = pd.read_excel(f'dados pca/{arq}')
        aux = aux.loc[aux['Language'] == 'English']
        abstracts.extend(df['Abstract'].tolist())
```

```python
# Seleciona aleatoriamente três abstracts da lista
random_abstracts = random.sample(abstracts, 3)

random_abstracts
```

```python
s3 = random_abstracts[0]
doc3 = nlp(s3)
```

```python
doc3_ns = [t for t in doc3 if not t.is_stop]
```

```python
from spacy import displacy

displacy.render(doc3, style='dep', jupyter=True)
```
