```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
arqs = os.listdir('aula 1 - dados')
```

```python
for i,arq in enumerate(arqs):
    
    if i == 0:
        df = pd.read_excel(f'aula 1 - dados/{arq}')
        df = df.loc[df['Language'] == 'English']
    else:
        aux = pd.read_excel(f'aula 1 - dados/{arq}')
        aux = aux.loc[aux['Language'] == 'English']
        df = pd.concat([df,aux])
```

```python
df.columns
```

```python
plt.hist(df['Number of Pages'])
plt.title('Número de páginas')
plt.show()
```

```python
plt.hist(df['Cited Reference Count'])
plt.title('Contagem de referências')
plt.show()
```
